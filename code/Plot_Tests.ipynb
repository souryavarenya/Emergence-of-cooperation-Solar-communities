```python
import csv
import numpy as np
import random

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from matplotlib.lines import Line2D
import matplotlib.animation as animation
```

```python
#Put Data into Arrays (TEST ONLY)

tag_list = []
x_coord_list = []
y_coord_list = []
idea_list = []
solartype_list = []

cnt = 0

with open('buildings_data.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        
        tag_list.append(row["building_code"])
        x_coord_list.append(float(row["building_coord_x"]))
        y_coord_list.append(float(row["building_coord_y"]))
        idea_list.append((random.random()*2)-1)
        solartype_list.append(random.randrange(0, 3, 1))
        
        cnt+=1
        
print("We have",cnt,"agents (buildings)")
```

```python
# COLOURED MAP PLOT FUNCTION

def ColourMap(x_axis, y_axis, col_axis, x_label="X_label", y_label="Y_label", colorbar=1, color_label=['Low', 'Medium', 'High'], title="Title", size=(15,10),cmap='RdYlGn',markersize=5,save=0,filename="test.svg"):

    #Configure Scatter Plot
    plt.figure(figsize=size)
    plt.title(title, fontsize=16)
    plt.xlabel(y_label, fontsize=8)
    plt.ylabel(x_label, fontsize=8)
    
    #Create Scatter Plot
    scatter = plt.scatter(x_axis,y_axis, c=col_axis, cmap=cmap, marker='s',s=markersize)
    
    #Setup Colorbar
    if(colorbar==1):
        min_col = min(col_axis)
        max_col = max(col_axis)
        cbar = plt.colorbar(ticks=[min_col, 0.5*(min_col+max_col), max_col])
        cbar.ax.set_yticklabels(color_label)
        
    #Alternatively, setup Legend
    else:
        
        viridis = cm.get_cmap(cmap, 3)
        
        legend_elements = [Line2D([0],[0], marker='s', color='w', label=color_label[0],markerfacecolor=viridis(0), markersize=15),
                   Line2D([0],[0], marker='s', color='w', label=color_label[1],markerfacecolor=viridis(1), markersize=15),
                   Line2D([0],[0], marker='s', color='w', label=color_label[2],markerfacecolor=viridis(2), markersize=15)]
        
        plt.legend(handles=legend_elements, loc='lower left')
    
    #Save image if requested
    if(save==1):
        plt.savefig(filename, format='svg')
```

```python
#Plot ColoredMap
ColourMap(x_coord_list,y_coord_list,idea_list,x_label="",y_label="",color_label=['Negative','Neutral','Positive'],cmap='RdYlGn',title="Initial Opinion Map", save=1)
```

```python
# Make 0=>Grey (Style choice)
cmap = 'YlGn'
viridis = cm.get_cmap(cmap, 3)
newcolors = viridis(np.linspace(0, 1, 3))
pink = np.array([128/256, 128/256, 128/256, 1])
newcolors[:1,:] = pink
newcmp = ListedColormap(newcolors)

#Plot ColoredMap
ColourMap(x_coord_list,y_coord_list,solartype_list,x_label="",y_label="",colorbar=0,color_label=['No Solar','Individual Solar','Solar Community'],cmap=newcmp,title="Final Solar Community Map")
```

```python
# Auxiliary function to create the Animation
def update_plot(i, data, scat):
    scat.set_array(data[i])
    return scat,

# COLOURED MAP ANIMATION (now the color vector is a Matrix, with all the color vectors to represent)

def AnimateColourMap(numframes, x_axis, y_axis, col_matrix, x_label="X_label", y_label="Y_label", colorbar=1, color_label=['Low', 'Medium', 'High'], title="Title", size=(15,10),cmap='RdYlGn',markersize=5,save=0,filename="test.svg"):

    #Configure Scatter Plot
    fig = plt.figure(figsize=size)
    plt.title(title, fontsize=16)
    plt.xlabel(y_label, fontsize=8)
    plt.ylabel(x_label, fontsize=8)
    
    #Create Scatter Plot
    scatter = plt.scatter(x_axis,y_axis, c=col_matrix[0], cmap=cmap, marker='s',s=markersize)
    
    #Setup Colorbar
    if(colorbar==1):
        min_col = min(col_matrix[0])
        max_col = max(col_matrix[0])
        cbar = plt.colorbar(ticks=[min_col, 0.5*(min_col+max_col), max_col])
        cbar.ax.set_yticklabels(color_label)
        
    #Alternatively, setup Legend
    else:
        
        viridis = cm.get_cmap(cmap, 3)
        
        legend_elements = [Line2D([0],[0], marker='s', color='w', label=color_label[0],markerfacecolor=viridis(0), markersize=15),
                   Line2D([0],[0], marker='s', color='w', label=color_label[1],markerfacecolor=viridis(1), markersize=15),
                   Line2D([0],[0], marker='s', color='w', label=color_label[2],markerfacecolor=viridis(2), markersize=15)]
        
        plt.legend(handles=legend_elements, loc='lower left')

    #Create animation and save .gif
    ani = animation.FuncAnimation(fig, update_plot, frames=range(numframes),fargs=(col_matrix, scatter))

    ani.save('test.gif', writer='imagemagick')

    plt.show()
```
